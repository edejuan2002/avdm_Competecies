```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Texto de la nube de palabras
text = "Dios PJaime PAlexis AndrésR Josemi Joaquín DavidN Jesús Martí Gerard MarcosB Adrián Víctor PabloT Óscar PauN Salva Álvaro PauM DaniB MarcosR Josep AndrésG Gonzalo JoseRa JoseA Bernat PabloG JorgeN"

# Generación de la nube de palabras con múltiples colores
wordcloud = WordCloud(width=800, height=400, colormap='viridis', background_color='white', repeat=False).generate(text)

# Mostrar la nube de palabras
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
