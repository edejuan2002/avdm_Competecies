```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Texto de la nube de palabras
text = "Dios PJaime PAlexis AndrsR Josemi Joaqun DavidN Jess Mart Gerard MarcosB Adrin Vctor PabloT scar PauN Salva lvaro PauM DaniB MarcosR Josep AndrsG Gonzalo JoseRa JoseA Bernat PabloG JorgeN"

# Generacin de la nube de palabras con mltiples colores
wordcloud = WordCloud(width=800, height=400, colormap='viridis', background_color='white', repeat=False).generate(text)

# Mostrar la nube de palabras
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
